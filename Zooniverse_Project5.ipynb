```python
#Importing all the necessary Goodies
import matplotlib.pyplot as plt
import numpy as np
from marvin.tools.maps import Maps
import marvin.utils.plot.map as mapplot
from marvin.tools.image import Image
#plt.style.use('seaborn-darkgrid')  # set matplotlib style sheet
from astropy.table import Table
import marvin

#pulling in the table for galaxies
t = Table.read('z_Ha_filt.fits')
```

```python
ifu_col = t['IFUDESIGN']
plt_ifu = t['PLATEIFU']
```

```python
len(ifu_col)
```

```python
from marvin.tools.image import Image
im = Image('8553-9102')
# generate a new image
# inputs are height and width in arcsec, and a arcsec/pixel scale
im.get_new_cutout(40, 40, scale=0.192)
# plot the new image cutout
ax = im.plot()
```

```python
n = len(ifu_col)
ifu_list = ifu_col.tolist()
bundle_size = []

#setting up a loop to pull out the bundle sizes for each IFU
for i in range(n):
    lst = ifu_list[i]
    lst = str(lst)
    a_case = lst.split('0')
    val = int(a_case[0])
    bundle_size.append(val)
    
#print(bundle_size)
```

```python
# 19, 37, 61, 91, 127
#12", 17", 22", 27", 32"
arc_sec = []
    

#beginning the loop for determining the image size in Arc Seconds
for i in range(n):
    #determining setting up bundle size
    if bundle_size[i] == 19:
        arc_sec.append(12)
    elif bundle_size[i] == 37:
        arc_sec.append(17)
    elif bundle_size[i] == 61:
        arc_sec.append(22)
    elif bundle_size[i] == 91:
        arc_sec.append(27)
    else:
        arc_sec.append(32)
        
    
#h_w = arc_sec + int(.5*arc_sec)
#scl_f = 7.68 / h_w
```

```python
#initializing lists
h_w = []
scl_f = []

#setting up the scale factor & the height/width for the images
for i in range(n):
    val = arc_sec[i] + int(.5 * arc_sec[i])
    scl_val = 7.68 / val
    h_w.append(val)
    scl_f.append(scl_val)
```

```python
#from marvin.tools.image import Image

for i in range(n):
    name=t["PLATEIFU"][i]
    # generate a new image
    im = Image(plt_ifu[i])
    # inputs are height and width in arcsec, and a arcsec/pixel scale
    im.get_new_cutout(40, 40, scale= .16)
    # plot the new image cutout
    ax = im.plot()
    # change the style of the hexagon
    im.overlay_hexagon(ax, color='cyan', linewidth=1)
    plt.show()
    
    plt.savefig(name+'IMG.png')
    print(i)
```
